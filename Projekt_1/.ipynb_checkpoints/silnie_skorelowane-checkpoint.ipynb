```python

import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('data/Android_Malware.csv', low_memory = False)
df.head()
```

```python
cor = df.corr()
```

```python
import math
df2= df.copy()
cor_list = []
for i in range(78):
    for j in range(78):
        if i < j and (abs(cor.iloc[i,j]) > 0.9 or math.isnan(cor.iloc[i,j])):
            cor_list.append(cor.columns[j])
```

```python
cor_list
```

```python
df2 = df2.drop(cor_list, axis = 1)
df2
```
